```python
dates=[]
costs=[]
aff_group=[]
country_name=[]
data=[]
sheet_name='Trovit_AT'
sheet = wb[sheet_name]
for row in sheet.iter_rows(min_row=3,max_row=408, min_col=1, max_col=7):
    for cell in row:
        if cell.column==1:
            datum = cell.value
        elif cell.column==7:
            cost = cell.value
            if datum != None:
                dates.append(datum)
                if cost !=None:
                    costs.append(cost)
                    aff_group.append(sheet_name)
                    country_name.append(country)
data=list(zip(dates,aff_group,country_name,costs))
print(data)
now=datetime.datetime.now()
cur_date=now.strftime("%Y%m%d")

with open(path + cur_date +'_output'+ '.csv', 'w') as csvfile:
    filewriter = csv.writer(csvfile, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    filewriter.writerow(['Date', 'AffiliateGroup', 'Country','Cost'])
    filewriter.writerow(data[0])
#fix how date is displayed and empty row
```
